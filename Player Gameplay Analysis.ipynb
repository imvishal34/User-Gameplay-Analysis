# Final Loyalty Point Formula

Loyalty Point=(0.01*Deposit Amount)+(0.005*Withdrawal Amount)+(0.001*Maximum of((No.of Deposit-No. of Withdrawal),0)+(0.2*No. of Games Played)

```python
df=df.rename(columns = {'Amount_x':'amt_dep','Amount_y':'amt_wit','Games Played':'games_played'})
df.head()
```

```python
df['loyalty_pts']=(0.01*df['amt_dep'])+(0.005*df['amt_wit'])+(0.2*df['games_played'])+(0.001*(df['dep_count']-df['wit_count']))
df.head()
```

```python
#Ranking based on the loyalty points
df["Rank"] = df["loyalty_pts"].rank(method='min',ascending=False)
df.head()
```

```python
#Here it is visible that some users have same ranks based on loyalty points
df.sort_values('Rank',ascending=False)
```

```python
#If the users have equal ranks then we will assign better rank to the user which has played more number of games
for user in df['User Id']:
    for i in range(1000):
        while(df['Rank'][user]==df['Rank'][i] and user!=i):
            if(df['games_played'][user]>df['games_played'][i]):
                df['Rank'][i]=df['Rank'][i]+1
            else:
                df['Rank'][user]=df['Rank'][user]+1       
```

```python
#These are the users with same number of loyalty points as well as equal number of games played
df[df.duplicated(['Rank'], keep=False)]
```

# Top 50 Players

```python
#These are the top ranked players based on Loyalty Points(1st Criteria) and Number of Games Played(2nd Criteria)
top_50=df.nlargest(50,'loyalty_pts')
top_50.head(50)
```

```python
sns.lineplot(x="User Id", y="Rank",
             data=top_50)
```

Now to allocate the pool of Rs.50000 we will be giving 50% Weightage to Loyalty Points, 30% Weightage to Number of Games Played and 20% Weightage to Amount Deposited by the user

```python
#Here we are assigning Bonus Points based on the weightage's assigned above
top_50['bonus_pts']=(0.5*df['loyalty_pts'])+(0.2*df['amt_dep'])+(0.3*df['games_played'])
top_50.head()
```

```python
Add= top_50['bonus_pts'].sum()
print(Add)
```

```python
#This way we will distribute the award money from the pool of Rs.50000 among the top 50 users
top_50['award_money']=(top_50['bonus_pts']/Add)*50000
top_50.head(50)
```

```python
sns.lineplot(x="Rank", y="award_money",
             data=top_50)
```
